```python
import numpy as np
import pandas as pd
```

```python
file_list = ['test_submit_0831_ch64_gram.csv','test_submit_0831_ch32_gram_2_best.csv',\
             'test_submit_0831_ch32.csv','test_submit_0831_3.csv']
new_df = pd.DataFrame()
new_list = []
for i, f in enumerate(file_list):
    test_file =  pd.read_csv(f)
    print ('length', len(test_file))
    print (test_file.keys())
    test_file.drop('ID', axis=1,inplace=True)
    col_names =  test_file.keys().tolist()
    test_matrix = test_file.values
    print ('test_matrix shape', test_matrix.shape)
    
    new_list.append(test_matrix)
    
new_matrix = np.array(new_list)
print (new_matrix.shape)
```

```python
mean_matrix = np.mean(new_matrix, axis=0)
mean_matrix.shape
```

```python
val_matrix = pd.DataFrame(mean_matrix)
val_matrix.columns = col_names
val_matrix.head()
```

```python
test_file = pd.read_csv ('test_submit_0831_ch64_gram.csv')


ensemble_df = pd.concat([test_file.ID,val_matrix],axis=1)
ensemble_df.head()
```

```python
ensemble_df.to_csv('test_submit_0901_ensemble_ch64_gram.csv',index=False)
```
